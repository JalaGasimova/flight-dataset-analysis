**4. DATA TRANSFORMATION**

*Numeric dÉyÉrlÉr arasÄ±ndaki kolinearlÄ±Äa baxÄ±rÄ±q. ÆgÉr vif dÉyÉri=1 olarsa,bu o demÉkdir ki kolinearlÄ±q yoxdur,vif= 1-5 arasÄ± olarsa, o demÉkdir ki zÉif/normal kolinearlÄ±q var. Vif dÉyÉri >5 olmasÄ± isÉ multikolinearlÄ±q probleminin olmasÄ±nÄ± gÃ¶stÉrir.*

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
x=df[['price','duration','stops']]
x=add_constant(x)
vif=pd.DataFrame()
vif['Variable']=x.columns
vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
print(vif)
for i,row in vif.iterrows():
    if row['Variable']=='const':
        continue
    elif row['VIF']<=1:
        print(f"{row['Variable']}: KolinearlÄ±q yoxdur")
    elif row['VIF']>1 and row['VIF']<5:
        print(f"{row['Variable']}: Normal kolinearlÄ±q var")
    else:
        print(f"{row['Variable']}: MultikolinearlÄ±q problemi var")
```
